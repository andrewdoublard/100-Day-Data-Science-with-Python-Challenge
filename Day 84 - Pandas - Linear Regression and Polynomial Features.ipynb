## Exercise 1:
#### The following df DataFrame is given:

#### The first column describes the years of experience (feature variable), the second column describes the salary (target variable).



#### Using the scikit-learn package and the LinearRegression class, find the linear regression equation for this problem.

#### In response, print the result to the console as shown below.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


df = pd.DataFrame(
    {
        'years': [1, 2, 3, 4, 5, 6],
        'salary': [4000, 4250, 4500, 4750, 5000, 5250],
    }
)

lr = LinearRegression()

lr = lr.fit(X = df[['years']], y = df[['salary']])


print(
    f'Linear regression: {lr.intercept_[0]:.2f} + '
    f'{lr.coef_[0][0]:.2f}x'
)
```

## Exercise 2:
#### Load data.csv file into a DataFrame and assign to df variable.

#### Then, based on the variable column, create a linear regression model that allows you to predict the values of the target column (create the model based on all available data). Use the scikit-learn package and the LinearRegression class.

#### Evaluate the model using score() and print the result (to four decimal places) to the console as shown below.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


## Reading csv into dataframe

df = pd.read_csv('data.csv')

## Fitting the lr with the df

lr = LinearRegression()

lr = lr.fit(X = df[['variable']], y = df[['target']])

## Model Score

score = lr.score(X = df[['variable']], y = df['target'])

# Print the model score
print(f'LinearRegression Score = {score:.4f}')
```
